```python
import numpy as np 
import Physical_Const as phys
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.optimize import brentq
from scipy.integrate import ode, odeint


import neutrinos as nunu
import CarbonFusion as eCC
import EoS_ionee as EOS
import ion_electron as ionee
import ions as ion
import electron_electron as ee
import Cv_star as cv


#----Constants------

hbar     = phys.hbar
c        = phys.c
G        = phys.G
SigmaSB    = phys.sigmaSB
me_mev   = phys.me
mp_mev   = phys.mp
e        = phys.e
e_erg    = phys.e_erg
mevtoerg = phys.mevtoerg
mu       = phys.mu 
kappaB   = phys.kappa
Msun     = phys.Msun
Rsun     = phys.Rsun

h        = hbar*(2.0*np.pi)
c2       = c*c
me       = me_mev * mevtoerg / c2
mp       = mp_mev * mevtoerg / c2

arad     = 4.0*SigmaSB / c
Tr       = me * c2 / kappaB
pi05     = np.sqrt(np.pi)
RBohr    = hbar**2 / (me*e_erg)
Lsun     = 3.846e33
yr       = (364)*24*60*60
avo      = 1.0/mu


# ---Dimensions---
MA       = 2.0
Cgrav    = G * Msun / c**2
mpl      = np.sqrt(hbar * c / G) # planck mass
Rdim     = np.power( mpl / me , 2.0) * hbar / (mpl * c)
Sigma02  = np.power( c / Rdim,2.0) / G

Sigma    = np.pi * me * np.power(me*c / h ,3.0) 
SigmaP   = c**2 * Sigma
rhodim   = ( ( mu * MA / me ) * 8.0* Sigma / 3.0 )
Tdim     = ( MA * mu *c**2) / kappaB
Ldim     = 4.0 * np.pi * Rdim**2 * SigmaSB * Tdim**4
kappaK   = (4.34e24*Sigma02*Rdim)*Sigma02*np.power(Tdim,-3.5)
#Sigma02  = ( mu * MA / me ) * 8.0* Sigma / 3.0
#Rdim     = c / np.sqrt( Sigma02 * G )
Mdim     = Rdim / Cgrav

Bdim     = (G / c**3 ) / Sigma02 


Jdim     = G / c**3
Jadim    = 1e50
JJdim    = Jadim * Jdim  / Rdim**2

#Tdim     = np.power(c**5 / ( 16.0**2 * np.pi**2 * G * sigma * Rsun**2) ,1.0/4.0)

rhobeta  = np.log10(3.9e10)
```

```python
#---degenerate pressure-------------
def Pch_EoS(x):
    x2  = x*x
    
    if x>1e-3:
        ww  = np.sqrt( 1.0 + x2) + x 
        pWD = x * np.sqrt( 1.0 + x2 ) * ( 2.0 * x2/3.0  - 1.0 ) +  np.log( ww ) 
    else:
        x3 = x2*x
        x5 = x3*x2
        x7 = x5*x2
        x9 = x7*x2
        x11 = x9*x2
        pWD = -2.77556e-17*x3 + 0.533333*x5 - 0.190476*x7 + 0.111111*x9 - 0.0757576*x11
    
    return SigmaP *pWD

def DpDe_ch(x):
    Dpdx = SigmaP * 8.0 * x**4 / ( 3.0 * np.sqrt( 1.0 + x**2 ) )
    Dedx = 3.0 * rhodim * x**2
    return Dedx/Dpdx
    
def func(x):
    return  rhodim *np.power(x,3)


NN      = 500
log_xXe = np.linspace( -5.0, 1.8,NN)
xx_xe   = np.power(10.0,log_xXe)

P_RFMT   = np.zeros(NN)
rho_RFMT = np.zeros(NN)

#file_EoS= " "
#rhoEoS,xeEoS,PEoS,PsatEoS,chip_EoS=np.loadtxt(file_EoS,usecols=(0,1,2,3,4),unpack=True)

#Create grid to fit fuction and calculate denrivatives EoS-------
for i in range(NN):
    xi          = xx_xe[i]
    P_RFMT[i]   = Pch_EoS(xi) / (c**2*Sigma02)
    rho_RFMT[i] = func(xi)  / Sigma02

EoS_RFMT    = interp1d( rho_RFMT , P_RFMT , bounds_error=False)
EoS_RFMT_02 = interp1d( P_RFMT , rho_RFMT , bounds_error=False)


DeDp = np.zeros(NN)
DxeF = 0.01
for i in range(NN):
    
    if i==0:
        Der = ( rho_RFMT[i+2] - rho_RFMT[0] ) / (  P_RFMT[i+2] - P_RFMT[0] )
    elif i== NN-1:
        Der = ( rho_RFMT[i] - rho_RFMT[i-2] ) / (  P_RFMT[i] - P_RFMT[i-2] )
    else:
        Der = ( rho_RFMT[i+1] - rho_RFMT[i-1] ) / (  P_RFMT[i+1] - P_RFMT[i-1] )
    
    DeDp[i] = Der

DeDp2 = c**2*DpDe_ch(xx_xe)

DeDpF2 = interp1d( P_RFMT , DeDp , bounds_error=False)
DeDpF = interp1d( P_RFMT , DeDp2 , bounds_error=False)

plt.loglog( Sigma02 * rho_RFMT, Sigma02*c**2*P_RFMT,'k:')
plt.loglog( Sigma02 * rho_RFMT, Sigma02*c**2*(rho_RFMT*(1e4/Tdim)),'r:')
plt.xlabel("rho")
plt.ylabel("P")
plt.show()

```

```python
def TOV(r,y):
    mns   = y[0]
    pns   = y[1]
    nuns  = y[2]
    rhons =  EoS_RFMT_02(pns)

    r2   = r*r
    r3   = r2*r
    ec2  = 2.0 * ( 4.0 * np.pi * r3 * pns + mns ) / ( r * ( r - 2.0 * mns ) )
    ec1  = 4.0 * np.pi * r2 *rhons
    ec3  = -0.5 * ec2 * ( pns + rhons )
    
    return np.array( [ec1,ec3,ec2])
```

```python
def Magnetic(rho, Bs, B0):
    x       = rho*Sigma02/1e9
    eta     = 0.8
    gamma   = 0.9
    Bfield  = Bs + B0*( 1.0-np.exp(-eta*x**gamma) )
    DBfield = B0 *  np.exp(-eta*x**gamma)  * eta*gamma*x**(gamma-1)
    return  [ Bfield, DBfield ]


def Newton(r,y, Lwd, Bs,B0):
    mns    = y[0]
    pns    = y[1]
    Ts     = y[2]
    rhons  = EoS_RFMT_02(pns)
    Pideal = Ts*rhons
    BB     = Magnetic(rhons, Bs, B0)
    #PB    = BB**2 /( 8.0*np.pi )
    rhoB   = BB[0]**2 / ( 8.0*np.pi*c**2 )
    drhodP = DeDpF(pns)
    kappa0 = kappaK * rhons**2* np.power(Ts,-3.5)
    
    r2   = r*r
    r3   = r2*r
    ec1  = 4.0 * np.pi * r2 * ( rhons  )
    ec2  =  mns  /  r2 
    gamma = (1.0/drhodP)*(pns/rhons)
    ec4a = (1.0 - 1.0/gamma) * ( ec2 * rhons  )
    if pns>Pideal:
        ec4 = 0.0
    else:
        ec4  = -( 3.0 / 16.0 ) * ( kappa0  / Ts**3 ) * ( Lwd / r2  )
    # ec3  = ( - ec2 * ( rhons + rhoB)  - beta*rhons*DTdr) / ( 1.0 +  BB[0] * BB[1] * drhodP/(4.0*np.pi) + beta *Ts * drhodP )
    ec3  = (- ec2 * ( rhons ) - rhons*ec4) / ( 1.0 + Ts*drhodP)
    
    return np.array( [ec1,ec3, ec4])
```

```python
Static = ode(Newton).set_integrator('dopri5')

#----Loop to calculate Static sequence--------

def StaticSeq(y0,r0, drr,rf, Lwd):
    global pwd, Twd, rhowd, rr
    #r0 = drr*1e-2
    Bs = 0.0*1e6 / Bdim
    B0 = 0.0*1e9 / Bdim

    pwd   = [ ]
    Twd   = [  ]
    rhowd = [ ]
    rr    = [ ]
   
    Static.set_initial_value(y0,r0).set_f_params(Lwd,Bs,B0)
    while Static.successful() and Static.t>rf and EoS_RFMT_02(Static.y[1])*Sigma02 > 5e-9 and Static.y[1]>0.0:
        mstar=Static.y[0]

        pwd.append( Static.y[1] )
        Twd.append( Static.y[2] )
        rhowd.append( EoS_RFMT_02(Static.y[1]) )
        rr.append( Static.t )
      
        rstar=Static.t

        EQs = Newton(Static.t,Static.y, Lwd, Bs,B0)
        data = Static.y
        #nustar = np.log( 1.0 - 2.0 * Static.y[0] / Static.t ) - Static.y[2]
        Static.integrate(Static.t+drr) 

        
        
    #print( Static.y, data, EQs )
    
    return [mstar,rstar]
```

```python
def WDRadius(Tc, Rwd, Lwd):
    r0 = 1e-5
    dr = r0/10
    
    xr = np.log10(10.0)
    xl = np.log10(0.01)
    xc = 0.5*(xr+xl)
    Ts = Tc

    Rl = StaticSeq( [ 0.0, np.power(10,xl), Ts ] ,r0, dr, 0.0, Lwd / Ldim )[1]
    Rr = StaticSeq( [ 0.0, np.power(10,xr), Ts ] ,r0, dr, 0.0, Lwd / Ldim )[1]
    print( Rl - Rwd/Rdim,  Rr- Rwd/Rdim)

    Rstar = StaticSeq( [ 0.0, np.power(10,xc), Ts ] ,r0, dr, 0.0, Lwd / Ldim )[1]
    while abs( Rstar - Rwd/Rdim)/ (Rwd/Rdim) > 1e-3:
        if Rstar- Rwd/Rdim > 0.0:
            xl = xc
        else:
            xr = xc
        xc = 0.5*(xl+xr)
        MM = StaticSeq( [ 0.0, np.power(10,xc), Ts], r0 , dr,0.0, Lwd/Ldim)
        Rstar = MM[1]
        
    return [xc, MM[0], MM[1] ]
```
